```python
# prompt: plot results of throws of dice
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

sides = 6
rolls = np.random.randint(1, sides + 1, 500)
bins = np.arange(0, sides + 1) + 0.5
plt.xkcd()
mpl.rcParams["font.family"] = ["DejaVu Sans"]

# Create a figure and axes
fig, ax = plt.subplots()
ax.hist(rolls, bins=bins, rwidth=0.7)

# Set the title and axes labels
ax.set_title("Throw a Dice")
ax.set_xlabel("Dice Roll")
ax.set_ylabel("How often?")

# Add a legend and show the plot
plt.show()
```

```python
# Generate poisson distribution
poisson = np.random.poisson(40, 360)
bins = np.arange(0, 80) + 0.5

# Create a figure and axes
fig, ax = plt.subplots()
ax.hist(poisson, bins=bins, rwidth=0.8)

# Set the title and axes labels
ax.set_title("Text messages per day, over last year")
ax.set_xlabel("messages in a day")
ax.set_ylabel("How many times?")

# Add a legend and show the plot
ax.legend()
plt.show()
```

```python
# Generate poisson distribution
me = np.random.poisson(4, 365)
myson1 = np.random.poisson(30, 365)
myson2 = np.random.poisson(50, 365)

bins = np.arange(0, 80) - 0.5

# Create a figure and axes
fig, ax = plt.subplots()
ax.hist(me, bins=bins, rwidth=0.8, color='red', label='me')
ax.hist(myson1, bins=bins, rwidth=0.8, color='blue', label='my son')
ax.hist(myson2, bins=bins, rwidth=0.8, color='green', label='my other son')

# Set the title and axes labels
ax.set_title("Text messages per day, last year")
ax.set_xlabel("messages per day")

# Add a legend and show the plot
ax.legend()
plt.show()
```
