## Imports and enviroment settings

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

plt.xkcd()
mpl.rcParams["font.family"] = ["Comic Sans MS"]
```

```python
# prompt: plot results of throws of dice

sides = 6
throws = 1000
rolls = np.random.randint(1, sides + 1, throws)
minx = rolls.min()
maxx = rolls.max()
bins = np.arange(minx, maxx + 2) - 0.5

# Create a figure and axes
fig, ax = plt.subplots()
ax.hist(rolls, bins=bins, rwidth=0.7)

# Set the title and axes labels
ax.set_title(f"Throw a Dice {throws} times")
ax.set_xlabel("Dice Roll")
ax.set_ylabel("How often?")

# Add a legend and show the plot
plt.show()
```

```python
# prompt: plot results of throws of dice

sides = 6
throws = 1000
roll1 = np.random.randint(1, sides + 1, throws)
roll2 = np.random.randint(1, sides + 1, throws)
rolls = roll1 + roll2

minx = rolls.min()
maxx = rolls.max()
bins = np.arange(minx, maxx + 2) - 0.5
# Create a figure and axes
fig, ax = plt.subplots()
ax.hist(rolls, bins=bins, rwidth=0.7)

# Set the title and axes labels
ax.set_title(f"Throw Two Dice {throws} times")
ax.set_xlabel("Dice Roll")
ax.set_ylabel("How often?")

# Add a legend and show the plot
plt.show()
```

```python
# Generate poisson distribution
days = 30
poisson = np.random.poisson(2, days)
bins = np.arange(0, 10) - 0.5

# Create a figure and axes
fig, ax = plt.subplots()
ax.hist(poisson, bins=bins, rwidth=0.8)

# Set the title and axes labels
ax.set_title(f"Text messages per day, (last {days} days)")
ax.set_xlabel("messages in a day")
ax.set_ylabel("How many days I got __ messages.")

# Add a legend and show the plot
plt.show()
```

```python
# Generate poisson distribution
days = 365
me = np.random.poisson(2, days)
myson = np.random.poisson(30, days)

bins = np.arange(0, 50) - 0.5

# Create a figure and axes
fig, ax = plt.subplots()
ax.hist(me, bins=bins, rwidth=0.8, color='red', label='me')
ax.hist(myson, bins=bins, rwidth=0.8, color='blue', label='my son')

# Set the title and axes labels
ax.set_title(f"Text messages per day, (last {days} days)")
ax.set_xlabel("messages in a day")
ax.set_ylabel("How many days I got __ messages.")
# Add a legend and show the plot
ax.legend()
plt.show()
```
